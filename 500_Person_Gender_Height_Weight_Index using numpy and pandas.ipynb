```python
# 7- What is the maximum and minimum height?
height= data[:,0]
Maxheight=height.max()
Minheight=height.min()
print('The maximum height= ',Maxheight, 'cm',',The minimum height = ',Minheight,'cm')
```

```python
# 8- What is the maximum and minimum weight?
weight=data[:,1]
Maxweight=weight.max()
Minweight=weight.min()
print('The max weight = ',Maxweight,'Kg',', The mininmum weight= ',Minweight,'kg' )
```

```python
# 9- Find the mean and standard deviation of height using the simple operations only. (np.sum) 
mean=height.sum()/height.size
print('Mean :',mean)
```

```python
diff=height-mean
squered=np.power(diff,2)
Sum=np.sum(squered)
n=height.size
div=Sum/(n-1)
standard_deviation=np.sqrt(div)
print('standard deviation :',standard_deviation )
```

```python
# 10- Find the mean and standard deviation of weight using the simple operations only.
mean=weight.sum()/weight.shape[0]
mean
```

```python
dif=weight-mean
powr=np.power(dif,2)
summ=np.sum(powr)
n=weight.size
div=summ/(n-1)
standard_deviation=np.sqrt(div)
print('standard deviation :', standard_deviation)
```

```python
# 11- Find the mean and standard deviation of height.
MeanOFHieght=np.mean(height)
StanderdDev=np.std(height)
print('The mean :',MeanOFHieght,'The standerd divaition',StanderdDev)
```

```python
# 12- Find the mean and standard deviation of weight..
MeanOFweight=np.mean(weight )
stdofweight=np.std(weight)
print('The mean :' ,MeanOFweight,'The std :',stdofweight)
```

```python
# 15- Normalize height and weight
Norofheight=(height-MeanOFHieght)/StanderdDev
NorOFweight=(weight-MeanOFweight)/stdofweight
print('Normalize height and weight : ',Norofheight,NorOFweight)
```
